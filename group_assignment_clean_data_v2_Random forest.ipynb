```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import FunctionTransformer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
y = df_dropna[['Churn']].values
df_dropna= df_dropna.drop(['Churn'], axis = 1)
df_dropna
```

```python
# Pipeline 
categorical_vars = list(df_dropna.select_dtypes(include=['object']).columns)

discrete_vars = list(df_dropna.select_dtypes(include=['int64']).columns)

continuous_vars = list(df_dropna.select_dtypes(include=['float64']).columns)

dep_var = ['Churn']
```

With so few missing values, and with these missing values balanced, it made more sense to drop them than to impute them. So there will be no imputation for missing values. But how should we scale discrete and continue variables?

```python
df_dropna.loc[:, discrete_vars + continuous_vars].hist(bins=50, figsize=(20,15))
plt.show()
# Since there are so many outliers, standardization will be better suited than normalization
```

```python
# Categorical - one hot encode
cat_ohe_step = ('ohe', OneHotEncoder(sparse=False, handle_unknown='ignore'))
cat_steps = [cat_ohe_step]
cat_pipe = Pipeline(cat_steps)
cat_transformers = [('cat', cat_pipe, categorical_vars)]

# Numerical - scale
num_scl_step = ('scl', StandardScaler())
num_steps = [num_scl_step]
num_pipe = Pipeline(num_steps)
num_transformers = [('num', num_pipe, discrete_vars + continuous_vars)]
```

```python
ct = ColumnTransformer(transformers=cat_transformers + num_transformers)
ct.fit(df_dropna[categorical_vars + discrete_vars + continuous_vars])
X = ct.transform(df_dropna[categorical_vars + discrete_vars + continuous_vars])

#y = df_dropna[['Churn']].values
```
